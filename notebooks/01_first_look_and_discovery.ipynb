## 2. Discover a real truth yourself

The safest, most satisfying early wins are **things you find in the data** — un-leakable, and they *are* the core lesson. Run the three cells below. Each is ~10 lines of pandas and each overturns a common SEO belief.

Every number is **computed live from the shipped CSV** — nothing is hardcoded.

```python
import pandas as pd, numpy as np
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")
print(df.shape[0], "rows,", df.shape[1], "columns")
df.head(3)
```

```python
df["trend_direction"].unique()
```

### Discovery A — "High search volume means more traffic." Does it?

```python
corr = df["search_volume"].corr(df["impressions_90d"])
print(f"Correlation between search_volume and impressions_90d: {corr:.3f}")
print("Near zero -> keyword search volume barely predicts the traffic a page actually gets.")
```

### Discovery B — the CTR cliff by position
Click-through rate is not flat: it collapses as you move down the results.

```python
visible = df[df["impressions_90d"] >= 100]
ctr_by_pos = visible.groupby("position_tier")["ctr"].mean().sort_values(ascending=False)
print(ctr_by_pos.round(4).to_string())
ctr_by_pos.plot(kind="bar", title="Mean CTR by position tier (impressions >= 100)", ylabel="CTR");
```

### Discovery C — is longer content the lever?
Compare word count for **declining** vs **growing** pages.

```python
wc = df.groupby("trend_direction")["word_count"].median()
print(wc.round(0).to_string())
print("\n'down' vs 'up' pages have almost the same median word count -> length is not the lever.")
```

```python
df_filtered_by_impressions = df[df["impressions_90d"] > 0]
corr = df_filtered_by_impressions["search_volume"].corr(df_filtered_by_impressions["impressions_90d"])
print(f"Correlation between search_volume and impressions_90d: {corr:.3f}")
print("Near zero -> keyword search volume STILL barely predicts the traffic a page actually gets.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Group by position and content type
ctr_analysis = df.groupby(['position_tier', 'content_type'])['ctr'].mean().unstack()

print("Mean CTR by Position Tier and Content Type:")
display(ctr_analysis.round(4))

# Visualize it to see which content_type is consistently at the bottom of the bars
ctr_long = df.groupby(['position_tier', 'content_type'])['ctr'].mean().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(data=ctr_long, x='position_tier', y='ctr', hue='content_type')
plt.title('CTR by Content Type across Position Tiers')
plt.ylabel('Mean CTR')
plt.show()
print("I observed comparison articles usually have the worst CTR!")
```

```python
# Map trend directions to a numeric scale
trend_map = {
    'down': -1,
    'flat': 0,
    'stable': 0,
    'new': 0,
    'up': 1
}

df['trend_score'] = df['trend_direction'].map(trend_map)

# Now calculate correlation with content age
corr = df['trend_score'].corr(df['content_age_days'])

print(f"Correlation between content_age_days and trend_score: {corr:.3f}")
print("Observation: Age and trend direction are somewhat linked in this dataset.")
```
